## Problem 1

A fair six-sided die is rolled twice. What is the conditional
probability that the first roll showed a 2, given that the sum of the
two rolls is 6?

1.  $\frac{1}{6}$

2.  $\frac{1}{5}$

3.  $\frac{3}{11}$

4.  $\frac{2}{5}$

**Correct answers:** (b)

**Explanation:** (B) $\frac{1}{5}$. There are five equally likely ways
for two die to sum to 6: (1,5), (2,4), (3,3), (4,2), (5,1). Among them,
one option (2,4) had a first roll of 2. Therefore the conditional
probability that the first roll showed a 2, given that the sum of the
two rolls is 6, is $\frac{1}{5}$.

## Problem 18

1.  Consider a fully-connected neural network with 3 input neurons, 4
    hidden neurons, and 2 output neurons. What is the total number of
    parameters, including bias units for non-input layers?

<!-- -->

1.  9

2.  11

3.  24

4.  26

**Correct answers:** (d)

**Explanation:** From the input layer to the hidden layer, you have 3
neurons fully connected to 4 neurons, which gives us $3 \cdot 4 = 12$
weights. Plus, there are 4 neurons in the hidden layer, which means
there are 4 biases. So, for the first connection, there are $12+4 = 16$
parameters. From the hidden layer to the output layer, you have 4
neurons fully connected to 2 neurons, which gives us $4 \cdot 2 = 8$
weights. Plus, there are 2 neurons in the output layer, which means
there are 2 biases. So, for the second connection, there are $8+2 = 10$
parameters. In total, we have $16+10 = 26$ parameters.
